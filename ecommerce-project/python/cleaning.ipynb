```python
import numpy as pn
import pandas as pd
```

Load Dataset

```python

df = pd.read_csv(r"C:\Users\Keshav\vs code folder\data-analytic-projects\ecommerce-project\data\raw_data.csv")
```

Standardize column name

```python

df.columns = df.columns.str.strip().str.lower()
df.info()
```

Handle Missing Values

```python
# Columns:

# Item
df["item"] = df["item"].fillna("Not available")

# Price Per Unit
df["price per unit"] = pd.to_numeric(df["price per unit"], errors="coerce")
df["price per unit"] = df["price per unit"].fillna(df["price per unit"].median()).astype(float)

# Quantity
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
df["quantity"] = df["quantity"].fillna(1).astype(int)

# Discount Applied
df["discount applied"] = df["discount applied"].fillna("False")
df["discount applied"] = df["discount applied"].replace({ 
    "True":True,
    "False":False,
    True:True,
    False:False
})

df.info()
df.isnull().sum()
df.head()
```

Fix Data Types

```python
# Convert:

# Date → datetime
df["transaction date"] = pd.to_datetime(df["transaction date"], errors="coerce")
df["transaction date"] = df["transaction date"].fillna(method="ffill")

df.info()
df.isnull().sum()
```

Total Spent 


```python
# Total Spent = Price × Quantity
df["total spent"] = df["price per unit"] * df["quantity"]

"Total Spent :",df["total spent"].sum()
```

Standardize Text

```python
# Fix:

# Category
df["category"] = df["category"].str.strip().str.title()

# Payment Method
df["payment method"] = df["payment method"].str.strip().str.title()

# Location
df["location"] = df["location"].str.strip().str.title()
```

Check Outliers

```python
df.describe()

# Outliers not present
```

Check Duplicates

```python
df[df.duplicated()]

df[df.duplicated(subset=['customer id', 'category', 'item', 'price per unit',
       'quantity', 'total spent', 'payment method', 'location',
       'transaction date', 'discount applied'])]
```

Save Clean Dataset


```python
df.to_csv("cleaned_data.csv",index=False)
```
